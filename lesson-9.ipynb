Так как метод https://new.vk.com/dev/likes.getList отдаёт по 100 результатов за 1 раз, то нам придётся использовать цикл while, чтобы получить все 117, а не 100, добавив в параметры запроса **offset**


````python
import requests

# для начала получим

#зададим значение переменной, которая будет отвечать за количество лайков, чтобы проходить по циклу
#больше 100, чтобы цикл не встал при первом проходе

count_likes = 101 


for offset in range(0,count_likes,100):
    
    r = requests.get('https://api.vk.com/method/likes.getList',params={'type': 'post',
                                                                       'owner_id':-38369814, #id сообщества с минусом, профиля - без
                                                                       'item_id':19543, #id объекта
                                                                       'filter':'likes', #получить только тех, кто лайкал
                                                                       'count':100, # 100 - максимум за 1 запрос 
                                                                       'offset':offset
                                                                      })
    response = r.json()['response']
    count_likes = response['count'] # присваиваем новое значение переменной, чтобы цикл выполнялся до тех пор, пока не выполнится условие
    users_list = response['users']
    
    #выводим на экран длину списка и сам список id
    print(len(users_list),users_list)
````

Как видите выше, мы получили все 117 ID пользователей, которые нажали "Мне нравится".

Давайте для удобства доработаем цикл, чтобы элементы двух списков (а у нас их может быть значительно больше) объединялись в один (http://pythonworld.ru/tipy-dannyx-v-python/spiski-list-funkcii-i-metody-spiskov.html). В этом нам поможет функция **list.extend(L)**
> **list.extend(L)** – расширяет список list, добавляя в конец все элементы списка L 

То есть, при каждом прохождении цикла нам нужно использовать этот метод. Дополнительно нам потребуется создать пустой список, в который будут попадать новые результаты при каждом проходе цикла.
